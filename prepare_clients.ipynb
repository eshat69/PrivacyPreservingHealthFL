# Imports Libraries

```python
import pandas as pd
import os
```

# Create Client Folder

```python
os.makedirs(
    "data/clients",
    exist_ok=True
)

print("Client directory ready.")
```

# Load Clean Datasets

```python
cleveland = pd.read_csv("data/processed/cleveland_clean.csv")
hungarian = pd.read_csv("data/processed/hungarian_clean.csv")
switzerland = pd.read_csv("data/processed/switzerland_clean.csv")
va = pd.read_csv("data/processed/va_clean.csv")

print("All Clean Datasets Loaded")
```

```python
print("Cleveland :", cleveland.shape)
print("Hungarian :", hungarian.shape)
print("Switzerland :", switzerland.shape)
print("VA :", va.shape)
```

# Save as Federated Clients 
---
Cleveland as        a
Hungarian as        b
Switzerland as      c
VA as               d
---

```python
cleveland.to_csv("data/clients/hospital_a.csv",index=False)
hungarian.to_csv("data/clients/hospital_b.csv",index=False)
switzerland.to_csv("data/clients/hospital_c.csv",index=False)
va.to_csv("data/clients/hospital_d.csv",index=False)

print("All client datasets saved.")
```

# Verify

```python
clients = [
    "hospital_a.csv",
    "hospital_b.csv",
    "hospital_c.csv",
    "hospital_d.csv"
]
for client in clients:
    df = pd.read_csv(f"data/clients/{client}")

    print(client,df.shape)
```

# Client Statistics

```python
client_files = [
    "hospital_a.csv",
    "hospital_b.csv",
    "hospital_c.csv",
    "hospital_d.csv"
]

for file in client_files:
    df = pd.read_csv(f"data/clients/{file}")
    print("\n" + "="*70)
    print(file)
    print(df["target"].value_counts())
    print(round(df["target"].value_counts(normalize=True)*100,2))
```

# Save Summary Table

```python
summary = []

for file in client_files:
    df = pd.read_csv(f"data/clients/{file}")

    summary.append({
        "Client": file,
        "Rows": len(df),
        "Disease %":
            round(df["target"].mean()*100,2)
    })

summary_df = pd.DataFrame(summary)
summary_df
```
